```python
#Explanation about escape sequence characters on python-
# representation  meaning
# \' 	Single Quote
# \\ 	Backslash
# \n 	New Line
# \r 	Carriage Return
# \t 	Tab
# \b 	Backspace
# \f 	Form Feed
# \ooo 	Octal value
# \xhh 	Hex value


print("Hello 'World'") #Hello 'World'
print('Hello \'World\'') #Hello 'World'

print("C:\\Users\\Documents") #C:\Users\Documents

print("Hello\nWorld") #Hello
                      #World

print("Hello\rWorld") #World
#\r carriage return : first it will print the "Hello" then \r will move the cursor to first line of the statement/letter , then again starts printing the next words after the \r which is "World" in this case - hence at the final stage we only able to see "World" as single string/word since it overwrites the existing string/word "Hello"
TestStr = "This is a test line.\r"
print(TestStr)  #This is a test line.
TestStr = "This is a test line.\rAnother line."
print(TestStr) #Another line.t line.



print("Hello\tWorld") #Hello    World

print("Hello\bWorld") #HellWorld : it presses the backspace/deletes previous word , then enters new word after the \b character


#\f form feed : '\f' in Python is the form-feed control character. It carries no intrinsic pagination behavior in Python itself; its effect depends on the consumer of the text.

# The octal value for 'H' is 110, 'e' is 105, 'l' is 154, 'o' is 157
print('\110\105\154\154\157')
# Output: Hello


print('\x48') #H
```
