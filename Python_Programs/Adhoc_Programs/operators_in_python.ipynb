```python
#operators in python -
#Operator precedence - PEMDAS (Parentheses â†’ Exponents â†’ Multiply/Divide â†’ Add/Subtract)
'''
1. Arithmetic Operators
2. Comparison Operators
3. Assignment Operators
4. Logical Operators
5. Bitwise Operators
6. Membership Operators
7. Identity Operators
'''

# 1. Arithmetic Operators
"""
Used for mathematical calculations.
| Operator | Meaning             | Example  | Result |
| -------- | ------------------- | -------- | ------ |
| `+`      | Addition            | `5 + 2`  | `7`    |
| `-`      | Subtraction         | `5 - 2`  | `3`    |
| `*`      | Multiplication      | `5 * 2`  | `10`   |
| `/`      | Division            | `5 / 2`  | `2.5`  |
| `//`     | Floor Division      | `5 // 2` | `2`    |
| `%`      | Modulus (remainder) | `5 % 2`  | `1`    |
| `**`     | Exponentiation      | `5 ** 2` | `25`   |
"""
def operations(operator:str,num1:int,num2:int)->None:
    match operator:
        case '+':
            print(num1+num2)
        case '-':
            print(num1-num2)
        case '*':
            print(num1*num2)
        case '/':
            print(num1/num2)
        case '//':
            print(num1//num2)
        case '%':
            print(num1%num2)
        case '**':
            print(num1**num2)
        case _:
            print("unknown operator")

num1:int = 20
num2:int = 2
operators = ['+','-','*','/','//','%','**','','$$','!','@']
for operator in operators:
    operations(operator,num1,num2)
```

```python
#2. Comparison Operators
"""
Used to compare values (return True or False).
| Operator | Meaning                  |
| -------- | ------------------------ |
| `==`     | Equal to                 |
| `!=`     | Not equal to             |
| `>`      | Greater than             |
| `<`      | Less than                |
| `>=`     | Greater than or equal to |
| `<=`     | Less than or equal to    |
"""
print(5==5)#True
print(2>5)#False
print(10.0<10)#False
print(11>=10)#True
print(12<=13) #True
print(1!=True) #False
print(1!=0) #True
print(1==True) #True
print('g'=='g') #True
print("h"!="g") #True
print("H"=="h") #False
```

```python
#3. Assignment Operators
"""
Used to assign values to variables.
| Operator | Example   | Same As               |
| -------- | --------- | --------------------- |
| `=`      | `x = 5`   | Assign 5 to x         |
| `+=`     | `x += 3`  | `x = x + 3`           |
| `-=`     | `x -= 3`  | `x = x - 3`           |
| `*=`     | `x *= 3`  | `x = x * 3`           |
| `/=`     | `x /= 3`  | `x = x / 3`           |
| `//=`    | `x //= 3` | Floor division assign |
| `%=`     | `x %= 3`  | Modulus assign        |
| `**=`    | `x **= 3` | Exponent assign       |
"""
num:int = 50
print(num) # 50 is assigned to num variable with = operator

num += 5
print(num) #55

num -= 3
print(num) #52

num *= 2
print(num) #104

num /= 4
print(num) #26.0

num //= 2
print(num) #13.0

num %= 2
print(num) #1.0

num **= 3
print(num) #1.0
```

```python
#4. Logical Operators
"""
Used to combine conditional statements.
| Operator | Meaning                                |
| -------- | -------------------------------------- |
| `and`    | True if both conditions are true       |
| `or`     | True if at least one condition is true |
| `not`    | Reverses the result                    |
"""
print(True and False)   # False
print(True or False)    # True
print(not True)         # False
```

```python
#5 Bitwise Operators
"""
Operate on binary numbers.
| Operator | Name        |
| -------- | ----------- |
| `&`      | AND         |
| `|'      | OR          |
| `^`      | XOR         |
| `~`      | NOT         |
| `<<`     | Left shift  |
| `>>`     | Right shift |
"""
#6 - 0110
#3 - 0011
print(6 & 3) #2(0010)
print(6 | 3) #7(0111)
print(6 ^ 3) #5(0101)
print(~6) #-(6+1) #-7
print(6 << 3) # equivalent to 6 * (2 power 3) = 6*(2**3), #48
print(6 >> 3) # equivalent to floor division 6//(2 power 3) =6//(2**3) , #0
```

```python
#6. Membership Operators
"""
Used to test if a value exists in a sequence.
| Operator | Meaning                              |
| -------- | ------------------------------------ |
| `in`     | Returns True if value exists         |
| `not in` | Returns True if value does not exist |
"""
print("a" in "apple")      # True
print("z" not in "apple")  # True
print("x" in "apple") #False
```

```python
#7. Identity Operators
"""
Used to compare memory locations.
| Operator | Meaning                           |
| -------- | --------------------------------- |
| `is`     | True if both refer to same object |
| `is not` | True if not same object           |
"""
x = [1, 2]
y = x
z = x.copy()
print(f"x-{id(x)} ,y-{id(y)} ,z-{id(z)}") #memory location
print(x is y)   # True
print(z is x)   # False
print(z is y)   # False
print(z is not x) #True
```
