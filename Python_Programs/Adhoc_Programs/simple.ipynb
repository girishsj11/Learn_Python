```python
from time import time
def execution_time(function):
    def wrapper(a:int,b:int):
        start = time()
        result = function(a,b)
        end = time()
        print(f"Total time taken for function to run is - {end-start} sec")
        return result
    return wrapper
```

```python
@execution_time
def addition(a:int,b:int) -> int:
    return a+b

@execution_time
def subtraction(a:int,b:int) -> int:
    return a-b

@execution_time
def multiplication(a:int,b:int) -> int:
    return a*b

@execution_time
def division(a:int,b:int) -> int:
    return a//b

a,b = 15,5
print(f"{addition.__name__} -- {addition(a,b)}")
print(f"{subtraction.__name__} -- {subtraction(a,b)}")
print(f"{multiplication.__name__} -- {multiplication(a,b)}")
print(f"{division.__name__} -- {division(a,b)}")
```

```python
class Employee:
    count: int = 0
    employer_name: str = "company"

    def __init__(self,name:str,age:int,group:str) -> None :
        self.name = name
        self.age = age
        self.group = group
        Employee.count += 1

    @classmethod
    def total_new_employee(cls) -> None:
        print(f"Total newly employees are added to {cls.employer_name} is {cls.count}")

    @staticmethod
    def leaves_assigned() -> None:
        print("In a every year 20 leaves will be allocated to each employee")

    def introduce(self) -> None:
        print(f"Welcome to new employee - {self.name} with age of - {self.age} & group - {self.group}")

Emp1:Employee = Employee("Giri",25,"Male")
Emp2:Employee = Employee("Aish",22,"Female")
Emp1.introduce()
Emp2.introduce()

Emp1.leaves_assigned()
Emp2.leaves_assigned()

Employee.total_new_employee()


```

```python
class Father:
    def __init__(self,father_name,father_age,father_salary):
        print("Initializing Father class")
        self.father_name = father_name
        self.father_age = father_age
        self.father_salary = father_salary

    def details(self):
        print(f"Fathers- '{self.father_name}' with age of '{self.father_age}' & earning salary of - '{self.father_salary}'")

class Son(Father):
    def __init__(self,father_name,father_age,father_salary,son_name,son_age):
        super().__init__(father_name,father_age,father_salary)
        print("Initializing Son class")
        self.son_name = son_name
        self.son_age = son_age

    def details(self):
        super().details()
        print(f"Son's- '{self.son_name}' with age of '{self.son_age}'")

S1:Son = Son("Father",70,25000,"Son",28)
S1.details()
```
