```python
#create dict comprehension to have key value pair , when key is even number then value of
#the key should be square , if its odd then it should be cubed.
numbers : dict = {num : (("Even",num**2) if num%2==0 else ("Odd",num**3)) for num in range(10)}
print(numbers)
```

```python
#sort the dict with its value got from the exam for each students.
students_marks = {"Ankur":57,"Anirud":37,"Chaya":65,"Avinash":77,"Giri":92,"Bob":32,
                  "Zyda" : 82 , "Herald" : 100, "india" : 99 , "Aish" : 88
                  }
sorted_students_marks = {key:value for key,value in sorted(students_marks.items(),key=lambda item:item[1])}
sorted_students_names = {key:value for key,value in sorted(students_marks.items(),key=lambda item:item[0])}
print(f"Original dict - {students_marks} "
      f"\nsorted dict based on the marks - {sorted_students_marks}"
      f"\nsorted dict based on the student names - {sorted_students_names}")
```
